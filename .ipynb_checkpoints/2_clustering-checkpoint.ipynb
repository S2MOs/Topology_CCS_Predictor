```python
# Make a Boltzmann weighing of the data to determine relative weights for conformers
col_names = list(global_data.columns)
col_toreweight = col_names[2:]
other_cols = col_names[:2]
weighted_data = global_data[other_cols]

for col in col_toreweight:
    weighted_data[col] = global_data[col] * global_data["weight"]

weighted_data = weighted_data.groupby(weighted_data.can, sort=False).sum()
print(f"Data shape = {weighted_data.shape}")
```

### Preprocessing - remove columns that are correlated, or without variance

```python
# zero-variance columns
zero_std_cols = weighted_data.columns[weighted_data.std() < 10e-6]
weighted_data = weighted_data[weighted_data.columns.difference(zero_std_cols)]

# standardize
scaled_data = pd.DataFrame(
    scale(weighted_data), index=weighted_data.index, columns=weighted_data.columns
)
print(f"Data shape = {scaled_data.shape}")
```

```python
# drop linearly dependent columns (use qr decomposition in a rank revealing way)
tol = 1e-8  # numeric tolerance for 0
Q, R, P = qr(
    scaled_data, pivoting=True
)  # pivoting qr decomposition (may take long), sorted diagonal values in R
rank = sum(
    np.abs(np.diag(R)) > tol
)  # calculate the rank of the matrix using the R part
col_idx_to_keep = P[:rank]  # fetch the top features from the P pivot vector
scaled_data = scaled_data[scaled_data.columns[col_idx_to_keep]]
print(f"Data shape = {scaled_data.shape}")
```

```python
# drop highly correlated columns
scaled_data_corr = scaled_data.corr().abs()
upper = scaled_data_corr.where(
    np.triu(np.ones(scaled_data_corr.shape), k=1).astype(bool)
)
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
scaled_data = scaled_data.drop(to_drop, axis=1)

print(f"Data shape = {scaled_data.shape}")
```

### Dimensionality reduction

```python
dfs = {}

# First 2 principal components
dfs["pc2"] = pd.DataFrame(
    PCA(n_components=2).fit_transform(scaled_data),
    index=scaled_data.index,
    columns=["PC1", "PC2"],
)

# Umap embedding, first 2 dimensions
nneigh = int(np.sqrt(scaled_data.shape[1]))
dfs["umap2"] = pd.DataFrame(
    UMAP(
        n_components=2, n_neighbors=nneigh, random_state=np.random.RandomState(420)
    ).fit_transform(scaled_data),
    index=scaled_data.index,
    columns=["UMAP1", "UMAP2"],
)
```

```python
# explained variance plot for PCA

fig, ax = plt.subplots(figsize=(6, 3))

pc = PCA()
pc.fit(scaled_data)
expl_var = pc.explained_variance_ratio_.cumsum()
pd.Series(expl_var, index=range(1, scaled_data.shape[1] + 1)).loc[:50].plot(
    style=".-", title="Explained variance for PC"
)

ax.set_xlabel("Number of components")
ax.set_ylabel("Explained variance")
ax.set_facecolor("#F8F8FB")

# plt.savefig('FigureSI_Variance_PC.svg', format='svg', bbox_inches='tight', dpi=600)
```

```python
# PCA and UMAP visualization

f, ax = plt.subplots(1, 2, figsize=(10, 4.5))

sns.scatterplot(data=dfs["pc2"], x="PC1", y="PC2", alpha=0.5, ax=ax[0]).set_title(
    "PC projection"
)
sns.scatterplot(data=dfs["umap2"], x="UMAP1", y="UMAP2", alpha=0.5, ax=ax[1]).set_title(
    "UMAP projection"
)

ax[0].set_facecolor("#F8F8FB")
ax[1].set_facecolor("#F8F8FB")
```

```python
# Generate the UMAP and PC embedding by changing the number of components (between 3 and 15)
# These data will be used to make a silhouette plot
for j in np.arange(3, 15, 1):
    dfs[f"umap{j}"] = pd.DataFrame(
        UMAP(
            n_components=j, n_neighbors=nneigh, random_state=np.random.RandomState(420)
        ).fit_transform(scaled_data),
        index=scaled_data.index,
        columns=[f"UMAP{i}" for i in range(1, j + 1)],
    )
    dfs[f"pc{j}"] = pd.DataFrame(
        (PCA(n_components=j)).fit_transform(scaled_data),
        index=scaled_data.index,
        columns=[f"PC{i}" for i in range(1, j + 1)],
    )
```

```python
# Compare the silhouette score for all number of dimensions
# UMAP
fig, ax = plt.subplots(figsize=(8, 3))

for j in np.arange(3, 15, 1):
    method = f"umap{j}"

    silhouette_df = pd.DataFrame(columns=["N_clust", "silhouette_score"])

    for NCLS in np.arange(2, 20, 1):
        Z = {}
        Z[method] = linkage(dfs[method], method="ward")
        cls = {}
        cls[method] = fcluster(Z[method], NCLS, criterion="maxclust")
        silhouette = silhouette_score(dfs[method], cls[method])
        temp_dict = {"N_clust": [NCLS], "silhouette_score": [silhouette]}
        temp_df = pd.DataFrame(data=temp_dict)
        silhouette_df = pd.concat([silhouette_df, temp_df])

    sns.scatterplot(data=silhouette_df, x="N_clust", y="silhouette_score")
    sns.lineplot(
        data=silhouette_df, x="N_clust", y="silhouette_score", label=f"umap{j}"
    )

ax.set_xlabel("Number of clusters")
ax.set_ylabel("Silhouette index")

ax.set_ylim(0.2, 0.9)

ax.legend(bbox_to_anchor=(1.02, 1.0), loc="upper left", fontsize=9)
ax.set_facecolor("#F8F8FB")

# plt.savefig('FigureSI_Silhouette_UMAP.svg', format='svg', bbox_inches='tight', dpi=600)

# 6-7 clusters seems to be optimal, regardless of the number of dimensions
```

### Get molecules from each cluster

```python
# Read global data
global_data = pd.read_excel("global_data.xlsx", index_col="inchi")
global_data = global_data.drop(
    columns=["can", "Unnamed: 0", "inchi_key", "index", "stoichiometry"]
)

# Make a Boltzmann weighing of the global data
col_names = list(global_data.columns)
col_toreweight = col_names[2:]
other_cols = col_names[:2]
weighted_data = global_data[other_cols]

for col in col_toreweight:
    weighted_data[col] = global_data[col] * global_data["weight"]

weighted_data = weighted_data.groupby(weighted_data.index, sort=False).sum()
# print(weighted_data)
```

```python
# produce rdkit molecules
inchidf = pd.DataFrame()
inchidf["inchi"] = weighted_data.index
mols = pd.Series(inchidf.inchi.map(Chem.MolFromInchi), index=inchidf.index).to_frame(
    "mol"
)
# concat molecules and clusters
mols = pd.concat([cls_df, mols], axis=1).dropna()
```
